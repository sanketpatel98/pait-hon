```python
newdata = data['DISTANCE']
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit_transform(data['ORIGIN'])
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
x = newdata
y = data['CANCELLED'] 
x.fillna(x.mean(),inplace=True)
```

```python
xtr,xts,ytr,yts = train_test_split(x,y,test_size = 0.9, random_state=1)
xtr = xtr.to_frame()
ytr = ytr.to_frame()
xts = xts.to_frame()
yts = yts.to_frame()
```

```python
lr.fit(xtr,ytr)
```

```python
prediction = lr.predict(xts)
```

```python
from sklearn.metrics import classification_report
```

```python
classification_report(yts,prediction)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(yts,prediction)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(yts,prediction)
```
